```python
value_counts = df['Built Year'].value_counts()
value_counts = value_counts[(value_counts.index >= 1909) & (value_counts.index <= 1921)]
explode = [0.1 if year in value_counts.index else 0 for year in value_counts.index]
plt.pie(value_counts, explode=explode, labels=value_counts.index, autopct='%1.1f%%')
plt.title("Built Year Distribution (1909-1921)")
plt.show()
```
